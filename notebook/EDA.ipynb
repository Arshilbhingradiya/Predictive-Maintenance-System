```python
features = ['Air temperature [K]', 'Process temperature [K]', 'Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]']
target = 'Failure Type'
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you already have your data in a DataFrame called `data`

# 1. Correlation Heatmap


# 2. Pair Plot
sns.pairplot(data, vars=features, hue=target, palette="Set2")
plt.title("Pair Plot")
plt.show()

# 3. Box Plots
for feature in features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=data, x=target, y=feature, palette="Set2")
    plt.title(f"Box Plot of {feature} by {target}")
    plt.show()

# 4. Histograms
for feature in features:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=data, x=feature, hue=target, kde=True, palette="Set2", bins=30)
    plt.title(f"Histogram of {feature} by {target}")
    plt.show()

# 5. Scatter Plots
for i, feature1 in enumerate(features):
    for feature2 in features[i+1:]:
        plt.figure(figsize=(8, 6))
        sns.scatterplot(data=data, x=feature1, y=feature2, hue=target, palette="Set2")
        plt.title(f"Scatter Plot of {feature1} vs {feature2} by {target}")
        plt.show()
```
